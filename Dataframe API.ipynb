```python
import findspark
findspark.init()
from pyspark.sql import SparkSession
spark = SparkSession.builder.master("local[*]").getOrCreate()
```

```python
import pandas as pd
amazonDF = pd.read_csv("Amazon_Responded_Oct05.csv", error_bad_lines=False)
amazonDF = amazonDF.astype(str) # Converting pandas df to string first
amazonDF = spark.createDataFrame(amazonDF) #converting pandas df to pyspark df
```

```python
# Select only the columns we need
amazonDF = amazonDF.select(['tweet_created_at', 'user_screen_name', 'user_id_str'])
```

```python
amazonDF.show()
```

```python
#Editing dataframe in required format
from pyspark.sql.functions import concat, col, lit
from pyspark.sql.functions import substring


amazonDF2 = amazonDF.withColumn('tweet_created_at_year', amazonDF['tweet_created_at'].substr(27, 4))

amazonDF2.show()
```

```python
amazonDF2 = amazonDF2.withColumn('tweet_created_at',amazonDF2['tweet_created_at'].substr(5, 6))
amazonDF2 = amazonDF2.select(concat(col("tweet_created_at"),lit(" "),col("tweet_created_at_year")).alias("Date"),"user_screen_name","user_id_str")
amazonDF2.show()
```

## Task 1



Keeping the necessary coloumns

```python
amazonDF2.registerTempTable("amazonDF_temp")
daily_active_users = spark.sql("select user_screen_name, user_id_str, count(distinct Date) as date from amazonDF_temp group by user_screen_name,user_id_str order by date desc")
daily_active_users.show()
```

Filtering the active users 

```python
daily_active_users = daily_active_users.filter(daily_active_users.date>=5)
```

```python
daily_active_users.show()
```

Displaying the username and ID of active users

```python
daily_active_users = daily_active_users.select(daily_active_users.user_screen_name, daily_active_users.user_id_str)
daily_active_users.show()
```

## Task 2


Read the text data into a data frame 

```python
pathToFile = 'experiment.txt'
exptxt= spark.read.csv(pathToFile, inferSchema=True, header =True).toDF("user_id")
exptxt.show()
```

Joining daily_active_users and experiment.txt on user_id

```python
left_join = exptxt.join(daily_active_users, exptxt.user_id== daily_active_users.user_id_str,how='left') # Could also use 'left_outer'
left_join.show()
```

Forming a new coloumn Wheter_active

```python
#from pyspark.sql import functions as f
#df.withColumn('D', f.when(f.col('B') > 0, "Yes").otherwise("No")).show()
from pyspark.sql.functions import when
left_join = left_join.withColumn("Whether_active", when(left_join.user_id == left_join.user_id_str, "Yes").otherwise("No"))
experiment_user = left_join.select(left_join.user_id, left_join.Whether_active)
experiment_user.show()
```

Calculating the percentage of active users 

```python
total=experiment_user.count()
from pyspark.sql import functions as f
experiment_user.select(((f.sum(f.when(f.col('Whether_active') == 'Yes', 1))/(total))*100).alias('percentage')).show(truncate=False)
```

## Task 3


Reading the Amazon_Responded _Oct_05 csv into a data frame

```python
Tweets = pd.read_csv("Amazon_Responded_Oct05.csv", error_bad_lines=False)
Tweets = Tweets.astype(str) # Converting pandas df to string first
Amazon_Tweets = spark.createDataFrame(Tweets)
Amazon_Tweets.show(10, False) # False allows us to show entire content of the columns
```

Joining the three data frames 


```python
joinfinal = Amazon_Tweets.join(daily_active_users, Amazon_Tweets.user_id_str == daily_active_users.user_id_str).join(experiment_user, Amazon_Tweets.user_id_str == experiment_user.user_id)
joinfinal.show(100)
```

```python
joinfinal.count()
```

Exporting the dataframe to a Amazon_new csv file




```python
Amazon_new = joinfinal.toPandas()
Amazon_new.to_csv(r"Amazon_new.csv", index = False)
```
